```python
#Q2

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Departure=df["departure_time"].value_counts()
```

```python

bars=Departure.plot(kind="bar",xlabel="Departure Times",ylabel="Frequency",title="Departure Time Frequency")
plt.bar_label(bars.containers[0])
plt.show()
```

```python
Arrival=df["arrival_time"].value_counts()
```

```python
bars=Arrival.plot(kind="bar",xlabel="Arrival Times",ylabel="Frequency",title="Arrival Time Frequency")
plt.bar_label(bars.containers[0])
plt.show()
```

```python
#Q3

Source=df["source_city"].value_counts()
```

```python
bars=Source.plot(kind="bar",xlabel="Source Cities",ylabel="Frequency",title="Source City Frequency")
plt.bar_label(bars.containers[0])
plt.show()
```

```python
Destination = df["destination_city"].value_counts()
```

```python
bars=Destination.plot(kind="bar",xlabel="Destination Cities",ylabel="Frequency",title="Destination City Frequency")
plt.bar_label(bars.containers[0])
plt.show()
```

```python
#Q4

df["airline"].value_counts()
```

```python
plt.figure(figsize=(8,6))
sns.violinplot(x="airline",y="price",data=df,palette="Set3")
plt.xlabel("Airlines")
plt.ylabel("Price")
plt.title("Airline Price Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="airline",y="price",data=df,palette="Set3")
plt.xlabel("Airlines")
plt.ylabel("Price")
plt.title("Airline Price Distribution")
plt.show()
```

```python
#Q5

plt.figure(figsize=(8,6))
sns.boxplot(x="departure_time",y="price",data=df,palette="Set3")
plt.xlabel("Departure Times")
plt.ylabel("Price")
plt.title("Departure Time Price Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="arrival_time",y="price",data=df,palette="Set3")
plt.xlabel("Arrival Times")
plt.ylabel("Price")
plt.ylim(0,15000)
plt.title("Arrival Time Price Distribution")
plt.show()
```

```python
#Q6

plt.figure(figsize=(8,6))
sns.violinplot(x="source_city",y="price",data=df,palette="Set3")
plt.xlabel("Source City")
plt.ylabel("Price")
plt.title("Source Price Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.violinplot(x="destination_city",y="price",data=df,palette="Set3")
plt.xlabel("Destination City")
plt.ylabel("Price")
plt.title("Destination Price Distribution")
plt.show()
```

```python
#Q7

last_minute_deals = df[df['days_left'].isin([1, 2])]
plt.figure(figsize=(8, 6))
sns.boxplot(x='days_left', y='price', data=last_minute_deals, palette='Set3')
plt.xlabel('Days Left Before Departure')
plt.ylabel('Price')
plt.title('Price Distribution for Tickets Bought 1 or 2 Days Before Departure')
plt.show()
```

```python
#Q5 Redone in a simpler way showing the price variation between arrival and departure time

sns.relplot(x="departure_time",y="price",data=df,col="arrival_time",kind="line")
```

```python
#Q6 Redone in a simpler fashion showing the price variation with destination and source city

sns.relplot(x="source_city",y="price",data=df,col="destination_city",kind="line")
```

```python
#8

Air_price=df.groupby("airline")["price"].mean().sort_values(ascending=False)
print(Air_price)
```

```python
plt.plot(Air_price)
plt.xlabel("Airlines")
plt.ylabel("Average Price")
plt.title("Average Price of Airlines")
plt.show()
```
